#MATPLOTLIB AND SEABORN

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Dataset Creation

```
np.random.seed(42)

n = 300

data = {
    "Employee_ID": range(1001, 1001+n),

    "Gender": np.random.choice(
        ["Male", "Female"],
        n
    ),

    "Department": np.random.choice(
        ["IT", "HR", "Finance", "Marketing", "Operations"],
        n
    ),

    "Education": np.random.choice(
        ["Diploma", "Bachelor", "Master", "PhD"],
        n
    ),

    "Experience_Years": np.random.randint(
        1, 21, n
    ),

    "Age": np.random.randint(
        22, 60, n
    ),

    "Salary": np.random.randint(
        30000, 150000, n
    ),

    "Performance_Score": np.random.randint(
        50, 101, n
    ),

    "Work_Mode": np.random.choice(
        ["Remote", "Hybrid", "Onsite"],
        n
    ),

    "Satisfaction_Level": np.random.randint(
        1, 11, n
    )
}

df = pd.DataFrame(data)

df.to_csv("employee_dataset.csv", index=False)

print(df.head())
```

#Count Plot
Count employees by department.

What it shows
- Frequency of categories
- Which department has most employees
- Useful for workforce distribution

```
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Department"
)

plt.title("Employees per Department")

plt.show()
```

#Bar Plot
Average salary by department.

What it shows
- Compares mean salary
- Useful for compensation analysis
- Higher bars = higher average pay

```
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="Department",
    y="Salary"
)

plt.title("Average Salary by Department")

plt.show()
```

#Histogram
Salary distribution.

What it shows
- Shows frequency distribution
- Detects skewness
- Reveals salary ranges

```
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="Salary",
    bins=20,
    kde=True
)

plt.title("Salary Distribution")

plt.show()
```

#KDE Plot
Smooth salary distribution.

What it shows
- Smooth version of histogram
- Peaks indicate common salary ranges
- Useful for probability estimation

```
plt.figure(figsize=(8,5))

sns.kdeplot(
    data=df,
    x="Salary",
    fill=True
)

plt.title("Salary Density")

plt.show()
```

#Scatter Plot
Experience vs Salary.

What it shows
- Reveals relationships
- Positive trend = more experience → - higher salary
- Detects unusual employees

```
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Experience_Years",
    y="Salary"
)

plt.title("Experience vs Salary")

plt.show()
```

#Regression Plot
Experience vs Salary with trend line.

What it shows
- Shows correlation strength
- Trend line indicates expected salary growth

```
plt.figure(figsize=(8,5))

sns.regplot(
    data=df,
    x="Experience_Years",
    y="Salary"
)

plt.title("Regression Analysis")

plt.show()
```

#Box Plot
Salary by department.

What it shows
- Median
- Quartiles
- Outliers
- Salary variability

```
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Department",
    y="Salary"
)

plt.title("Salary Spread by Department")

plt.show()
```

#Violin Plot

Combines:

- Boxplot
- Density distribution

Wider sections indicate many employees in that salary range.

```
plt.figure(figsize=(8,5))

sns.violinplot(
    data=df,
    x="Department",
    y="Salary"
)

plt.title("Salary Distribution by Department")

plt.show()
```

#Strip Plot
Interpretation

Every dot = one employee.

Useful for:

- Raw observations
- Detecting clusters

```
plt.figure(figsize=(8,5))

sns.stripplot(
    data=df,
    x="Department",
    y="Salary"
)

plt.title("Individual Salaries")

plt.show()
```

#Swarm Plot
Interpretation

Like strip plot but avoids overlap.

```
plt.figure(figsize=(8,5))

sns.swarmplot(
    data=df,
    x="Department",
    y="Salary"
)

plt.title("Salary Swarm Plot")

plt.show()
```

#Heatmap (Most Important)
Correlation matrix.

Useful for:

- Feature selection
- Machine learning

```
plt.figure(figsize=(8,6))

corr = df.select_dtypes(
    include="number"
).corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

#Pair Plot (EDA Favorite)
Interpretation

Automatically generates:

- Scatter plots
- Histograms
- Relationships

Excellent for exploration.

```
sns.pairplot(
    df[
        [
            "Age",
            "Salary",
            "Experience_Years",
            "Performance_Score"
        ]
    ]
)

plt.show()
```

#Hue-Based Scatter Plot
Color by department.

Interpretation
- Compare groups visually
- Detect department-specific trends

```
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Experience_Years",
    y="Salary",
    hue="Department"
)

plt.title("Experience vs Salary by Department")

plt.show()
```

#Line Plot
Average salary by experience.

Interpretation

Shows trends across experience levels.

```
avg_salary = df.groupby(
    "Experience_Years"
)["Salary"].mean().reset_index()

sns.lineplot(
    data=avg_salary,
    x="Experience_Years",
    y="Salary"
)

plt.title("Average Salary Growth")

plt.show()
```

#FacetGrid
Split graphs by work mode.

Interpretation

Creates separate charts for:

- Remote
- Hybrid
- Onsite

Useful for segmented analysis.

```
g = sns.FacetGrid(
    df,
    col="Work_Mode"
)

g.map(
    sns.histplot,
    "Salary"
)

plt.show()
```

#Joint Plot
Interpretation

Contains:

- Scatter plot
- X distribution
- Y distribution

Great for correlation analysis.

```
sns.jointplot(
    data=df,
    x="Experience_Years",
    y="Salary",
    kind="scatter"
)

plt.show()
```
