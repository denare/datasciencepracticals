#numpy.linalg
Linear Algebra operations.

Function used:
- np.linalg.inv()
- np.linalg.det()
- np.linalg.eig()
- np.linalg.solve()
- np.linalg.norm()

```
import numpy as np

# Create a matrix A
A = np.array([
    [4, 2],
    [1, 3]
])

# Create vector b for solving equations
b = np.array([10, 8])

print("Matrix A:")
print(A)

print("\nVector b:")
print(b)
```

```
det_A = np.linalg.det(A)
print(det_A)
```

```
inverse_A = np.linalg.inv(A)
print(inverse_A)
```

```
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)
```

```
solution = np.linalg.solve(A, b)
print(solution)
```

```
norm_A = np.linalg.norm(A)
print(norm_A)
```

#numpy.fft
commonly used in signal processing, audio analysis, image processing, and communications.


Functions Used:
- np.fft.fft()- Converts a signal from the time domain to the frequency domain.
- np.fft.ifft()- Converts it back from the frequency domain to the time domain.

```
import numpy as np

# Original signal
signal = np.array([1, 2, 3, 4])

print("Original Signal:")
print(signal)

# Fast Fourier Transform
fft_result = np.fft.fft(signal)

print("\nFFT Result:")
print(fft_result)

# Inverse FFT
ifft_result = np.fft.ifft(fft_result)

print("\nRecovered Signal:")
print(ifft_result)
```

```
import numpy as np

# Simulated sensor readings
sensor_data = np.array([
    12, 15, 18, 20,
    22, 21, 19, 16
])

fft_data = np.fft.fft(sensor_data)

print("FFT:")
print(fft_data)

restored = np.fft.ifft(fft_data)

print("\nRestored Signal:")
print(restored)
```
