```
import pandas as pd
import numpy as np
```

```
data = {
    'Student_ID': range(1, 51),
    'Name': [f'Student_{i}' for i in range(1, 51)],
    'Age': np.random.randint(18, 30, 50),
    'Gender': np.random.choice(['Male', 'Female'], 50),
    'Course': np.random.choice(
        ['Cybersecurity', 'Data Science', 'Networking', 'Software Engineering'],
        50
    ),
    'GPA': np.round(np.random.uniform(2.0, 4.0, 50), 2),
    'Country': np.random.choice(
        ['Tanzania', 'Kenya', 'Uganda', 'Rwanda'],
        50
    ),
    'Graduated': np.random.choice([True, False], 50)
}
```

```
# Create DataFrame
df = pd.DataFrame(data)

# Display first 5 rows
print(df.head())

# Save as CSV
df.to_csv('students.csv', index=False)

# Save as Excel
df.to_excel('students.xlsx', index=False)

# Save as Parquet
df.to_parquet('students.parquet', index=False)

# Save as JSON
df.to_json('students.json', orient='records', indent=4)

print("Files saved successfully!")
```

# combine two dataset

```
import pandas as pd

students = pd.DataFrame({
    'Student_ID': [1, 2, 3, 4, 5],
    'Name': ['John', 'Mary', 'Peter', 'Alice', 'David'],
    'Age': [20, 21, 19, 22, 20],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male']
})

students.to_csv('students.csv', index=False)
print(students)
```

```
academic = pd.DataFrame({
    'Student_ID': [1, 2, 3, 4, 5],
    'Course': ['Cybersecurity', 'Networking', 'Data Science', 'Cybersecurity', 'Software Engineering'],
    'GPA': [3.8, 3.5, 3.9, 3.7, 3.6],
    'Year': [2, 3, 1, 4, 2]
})

academic.to_csv('academic_records.csv', index=False)
print(academic)
```

Use merge() when datasets share a common column.

```


students = pd.read_csv('students.csv')
academic = pd.read_csv('academic_records.csv')

merged_df = pd.merge(
    students,
    academic,
    on='Student_ID',
    how='inner'
)

print(merged_df)
```

Use concat() when rows correspond in the same order.

```
students = pd.read_csv('students.csv')
academic = pd.read_csv('academic_records.csv')

concat_df = pd.concat(
    [students, academic[['Course', 'GPA', 'Year']]],
    axis=1
)

print(concat_df)
```
